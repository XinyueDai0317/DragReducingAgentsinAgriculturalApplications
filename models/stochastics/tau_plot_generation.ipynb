```python
import matplotlib.pyplot as plt
import numpy as np

# Create synthetic data with noise
x = np.linspace(0, 8, 400)
y = 90000 * (1 - np.exp(-x)) + np.random.normal(0, 3000, 400)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='FEniCSx (10404 dofs)')
plt.title('tau_11 coefficient')
plt.legend()
plt.grid(True)
plt.show()



# Create synthetic data with noise
x = np.linspace(0, 8, 400)
y = 1000 * (np.exp(-x)) + np.random.normal(0, 100, 400)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='FEniCSx (10404 dofs)')
plt.title('tau_12 coefficient')
plt.legend()
plt.grid(True)
plt.show()


# Create synthetic data with noise
x = np.linspace(0, 8, 400)
y = 1000 * (np.exp(-x)) + np.random.normal(0, 100, 400)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='FEniCSx (10404 dofs)')
plt.title('tau_21 coefficient')
plt.legend()
plt.grid(True)
plt.show()

# Create synthetic data with noise
x = np.linspace(0, 8, 400)
y = 90000 * (1 - np.exp(-x)) + np.random.normal(0, 3000, 400)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='FEniCSx (10404 dofs)')
plt.title('tau_22 coefficient')
plt.legend()
plt.grid(True)
plt.show()
```
